##1b) MATPLOTLIB: Figsize / Rotate Label / X and y label / Title / Legend

##1c) PANDAS: Setting Title / X and Y Label / Legend in Pandas Plot

#**IS NOT AS EASY AS IT SEEMS!!!**

```python
#a dataframe must be loaded first
import pandas as pd

a = ['a', 'b', 'c', 'd', 'e']
b = [1,2,3,4,5]

df = pd.DataFrame(\
                 b,
                 index = a,
                 columns = ['col1']
                 )
df
```

```python
j = df.plot.line()

# Setting Title
j.set_title (\
             'BLABLABLA',
             fontsize = 21,
             y = 1.1
             # how much white space you want below the title
             )

# Setting Legend
j.legend(\
         loc = 4, #1,2,3,4
         fontsize = 21
         )

# Setting x and Y Label
j. set_xlabel(\
              'THE X AXIS',
              fontsize = 21)

j. set_ylabel(\
              'THE Y AXIS',
              fontsize = 21)
```

#Part 8) SUB PLOTS

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mtcars = pd.read_csv('https://www.alvinang.sg/s/mtcars.csv',)

mtcars
```

##8a) Sub Plot using Groupby

```python
mtcars.groupby('cyl')\
      .mean()\
      .plot(subplots = True,
      figsize = (5,26))

```

##8b) Sub Plot using Subplots

```python
import matplotlib.pyplot as plt

# Create a figure with 2 rows and 3 columns of subplots
fig, axes = plt.subplots(2, 3, figsize=(16, 8))

# Plot different columns of the mtcars_sample dataset on each subplot
mtcars.plot(x='mpg',
            y='hp',
            kind='scatter',
            ax=axes[0, 0],
            title='MPG vs. Horsepower')

mtcars.plot(x='wt',
            y='qsec',
            kind='scatter',
            ax=axes[0, 1],
            title='Weight vs. Quarter Mile Time')

mtcars['cyl']\
             .value_counts()\
             .plot(kind='bar',
                   ax=axes[0, 2],
                   title='Cylinder Distribution')

mtcars['gear']\
              .value_counts()\
              .plot(kind='bar',
                    ax=axes[1, 0],
                    title='Gear Distribution')

mtcars['am']\
            .value_counts()\
            .plot(kind='bar',
                  ax=axes[1, 1],
                  title='Transmission Type Distribution')

mtcars['vs']\
            .value_counts()\
            .plot(kind='bar',
                  ax=axes[1, 2],
                  title='Engine Shape Distribution')

plt.tight_layout()
plt.show()
```
